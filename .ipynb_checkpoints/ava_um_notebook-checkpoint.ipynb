```python
# Importa bibliotecas necessárias e inicializa uma conexão com um servidor do banco de dados postgre-sql
import psycopg2
import random
import csv
from datetime import datetime

conn = psycopg2.connect(
    dbname="ava1_db",
    user="ava1_user",
    password="ava1_password",
    host="localhost",
    port="5432"
)

cursor = conn.cursor()

CGH_data = "Datasets/Centrais_Geradoras_Hidrelétricas_CGH.csv"
UHE_data = "Datasets/Usinas_Hidrelétricas_UHE.csv"
```

```python
# Define a função que cria schemas e tables para o postgre
# para futura inserção de dados.

# removed:     CREATE SCHEMA IF NOT EXISTS public AUTHORIZATION ava1_user;
def criar_tabelas(cursor):
    tabelas_ddl = """/* ava_um_modeloLogico: */

    CREATE SCHEMA IF NOT EXISTS ava1_schema AUTHORIZATION ava1_user;
    SET search_path TO ava1_schema;

    DROP TABLE IF EXISTS ava1_schema."Usina" CASCADE;
    DROP TABLE IF EXISTS ava1_schema."Rios" CASCADE;
    DROP TABLE IF EXISTS ava1_schema."Proprietarios" CASCADE;
    DROP TABLE IF EXISTS ava1_schema."Municipios" CASCADE;
    DROP TABLE IF EXISTS ava1_schema."Estados" CASCADE;
    DROP TABLE IF EXISTS ava1_schema."Localizacoes" CASCADE;

    CREATE TABLE IF NOT EXISTS ava1_schema."Usina" (
        id SERIAL PRIMARY KEY,
        nome VARCHAR NOT NULL,
        potkw FLOAT,
        potfis FLOAT,
        iniciop DATE,
        atualizacao DATE,
        lat DOUBLE PRECISION ,
        long DOUBLE PRECISION ,
        estagio VARCHAR,
        id_proprietario INTEGER NULL,
        id_rio INTEGER NOT NULL,
        id_localizacao INTEGER NOT NULL
    );

    CREATE TABLE IF NOT EXISTS ava1_schema."Proprietarios" (
        proprietario VARCHAR,
        id_proprietario INTEGER PRIMARY KEY NOT NULL
    );

    CREATE TABLE IF NOT EXISTS ava1_schema."Municipios" (
        municipio VARCHAR NOT NULL,
        uf CHAR(2) NOT NULL,
        id_municipio SERIAL PRIMARY KEY
    );

    CREATE TABLE IF NOT EXISTS ava1_schema."Estados" (
        uf CHAR(2) PRIMARY KEY
    );

    CREATE TABLE IF NOT EXISTS ava1_schema."Rios" (
        rio VARCHAR NOT NULL,
        id_rio INTEGER PRIMARY KEY
    );

    CREATE TABLE IF NOT EXISTS ava1_schema."Localizacoes" (
        id_localização INTEGER PRIMARY KEY,
        longitude DOUBLE PRECISION,
        id_municipio1 INTEGER NOT NULL,
        id_municipio2 INTEGER,
        latitude DOUBLE PRECISION
    );

    ALTER TABLE IF EXISTS ava1_schema."Usina"
    ADD CONSTRAINT FK_Usina_Proprietario
    FOREIGN KEY (id_proprietario)
    REFERENCES ava1_schema."Proprietarios"(id_proprietario);

    ALTER TABLE IF EXISTS ava1_schema."Usina"
    ADD CONSTRAINT FK_Usina_Rio
    FOREIGN KEY (id_rio)
    REFERENCES ava1_schema."Rios"(id_rio);

    ALTER TABLE IF EXISTS ava1_schema."Usina"
    ADD CONSTRAINT FK_Usina_Localizacao
    FOREIGN KEY (id_localizacao)
    REFERENCES ava1_schema."Localizacoes"(id_localização);

    ALTER TABLE IF EXISTS ava1_schema."Municipios" ADD CONSTRAINT FK_Municipios_1
        FOREIGN KEY (uf)
        REFERENCES ava1_schema."Estados"(uf);
    
    ALTER TABLE IF EXISTS ava1_schema."Localizacoes"
    DROP CONSTRAINT IF EXISTS FK_Localizacoes_2;

    ALTER TABLE IF EXISTS ava1_schema."Localizacoes"
    ADD CONSTRAINT FK_Localizacoes_Mun1
        FOREIGN KEY (id_municipio1)
        REFERENCES ava1_schema."Municipios"(id_municipio);

    ALTER TABLE IF EXISTS ava1_schema."Localizacoes"
    ADD CONSTRAINT FK_Localizacoes_Mun2
        FOREIGN KEY (id_municipio2)
        REFERENCES ava1_schema."Municipios"(id_municipio);
    COMMIT;"""
    cursor.execute(tabelas_ddl)
    conn.commit()
```

```python
# Define a função que carrega as tabelas criadas com as informações dadas nos arquivos .csv do segundo argumento
def carregar_dados(cursor, filepath):
    counter_IDusina = 0
    id_rio = 0
    id_prop = 0
    id_local = 0
    id_mun1 = 0
    id_mun2 = 0
    cursor.execute("SET datestyle = 'ISO, DMY';")
    with open(filepath, 'r', encoding='utf-8') as csvfile:
        leitor = csv.reader(csvfile, delimiter=';')
        header = next(leitor)

        for linha in leitor:
            nome_usina = linha[0]
            nome_proprietario = linha[1]
            mun_1 = linha[2]
            uf_1 = linha[3]
            mun_2 = linha[4]
            uf_2 = linha[5]
            nome_rio = linha[6]
            potkw = linha[7].replace(',', '.')
            atualizacao = linha[8]
            if atualizacao == "00000000":
                atualizacao = None
            iniop = linha[9]
            if iniop == "00000000":
                iniop = None
            potfis = linha[10].replace(',', '.')
            id_usina = linha[11]
            if id_usina == '0':
                id_usina = random.random() + counter_IDusina + 1
                cursor.execute('SELECT id FROM ava1_schema."Usina"')
                ids_usados = [row[0] for row in cursor.fetchall()]
                while id_usina in ids_usados:
                    id_usina = random.random() + counter_IDusina + 1
            lat = linha[12].replace(',', '.')
            long = linha[13].replace(',', '.')
            estag_imp = linha[14]

            cursor.execute(
                """
                INSERT INTO ava1_schema."Estados" (uf)
                VALUES (%s)
                ON CONFLICT (uf) DO NOTHING;
                """,
                (uf_1,)
            )
            if uf_2:
                cursor.execute(
                    """
                    INSERT INTO ava1_schema."Estados" (uf)
                    VALUES (%s)
                    ON CONFLICT (uf) DO NOTHING;
                    """,
                    (uf_2,)
                )
            
            if id_local == 0:
                cursor.execute('SELECT COALESCE(MAX(id_localização), 0) FROM ava1_schema."Localizacoes"')
                ultimo_id = cursor.fetchone()[0]
                id_local = ultimo_id + 1
            else:
                id_local+=1
            
            cursor.execute(
                """
                    INSERT INTO ava1_schema."Municipios" (municipio, uf)
                    VALUES (%s, %s)
                    RETURNING id_municipio
                    """,
                    (mun_1, uf_1)
            )
            result = cursor.fetchone()
            if result:
                id_mun1 = result[0]
            
            if mun_2:
                cursor.execute(
                    """
                    INSERT INTO ava1_schema."Municipios" (municipio, uf)
                    VALUES (%s, %s)
                    RETURNING id_municipio
                    """,
                    (mun_2, uf_2)
                )
                result = cursor.fetchone()
                if result:
                    id_mun2 = result[0]
                cursor.execute(
                    """
                    INSERT INTO ava1_schema."Localizacoes" (id_localização, longitude, id_municipio1, id_municipio2, latitude)
                    VALUES (%s, %s, %s, %s, %s)
                    """,
                    (id_local, long, id_mun1, id_mun2, lat)
                )
            else:
                cursor.execute(
                    """
                    INSERT INTO ava1_schema."Localizacoes" (id_localização, longitude, id_municipio1, id_municipio2, latitude)
                    VALUES (%s, %s, %s, %s, %s)
                    """,
                    (id_local, long, id_mun1, None, lat)
                )

            if nome_proprietario:
                if id_prop == 0:
                    cursor.execute('SELECT COALESCE(MAX(id_proprietario), 0) FROM ava1_schema."Proprietarios"')
                    ultimo_id = cursor.fetchone()[0]
                    id_prop = ultimo_id + 1
                else:
                    id_prop +=1
                cursor.execute(
                """
                    INSERT INTO ava1_schema."Proprietarios" (proprietario, id_proprietario)
                    VALUES (%s, %s)
                    """,
                    (nome_proprietario, id_prop)
                )
            
            if id_rio == 0:
                cursor.execute('SELECT COALESCE(MAX(id_rio), 0) FROM ava1_schema."Rios"')
                ultimo_id = cursor.fetchone()[0]
                id_rio = ultimo_id + 1
            else:
                id_rio+=1
            cursor.execute(
                """
                    INSERT INTO ava1_schema."Rios" (rio, id_rio)
                    VALUES (%s, %s)
                    """,
                    (nome_rio, id_rio)
                )
            if id_prop == 0:
                id_prop = None
            cursor.execute(
                """
                INSERT INTO ava1_schema."Usina" (nome, potkw, potfis, iniciop, atualizacao, lat, long, estagio, id_proprietario, id_rio, id_localizacao)
                VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
                """,
                (nome_usina, potkw, potfis, iniop, atualizacao, lat, long, estag_imp, id_prop, id_rio, id_local)
            )
            id_prop = 0
    try:
        conn.commit()
    except psycopg2.Error as e:
        print("Ocorreu um erro: {e}")
        conn.rollback()
```

```python
def fazer_query(sql_query):
    try:
        conn = psycopg2.connect(
            dbname="ava1_db", user="ava1_user", password="ava1_password", host="localhost", port="5432"
        )
        cursor = conn.cursor()

        cursor.execute(sql_query)
        #colnames = [desc[0] for desc in cursor.description]
        results = cursor.fetchall()
        print(results)
    except Exception as e:
        print("Ocorreu um erro: {e}")
```

```python
# Buscam as empresas que indiretamente mais geram energia através de suas usinas
query_1 = """SELECT 
    p.proprietario,
    SUM(u.potfis) AS total_potfis
FROM ava1_schema."Usina" u
JOIN ava1_schema."Proprietarios" p ON u.id_proprietario = p.id_proprietario
GROUP BY p.proprietario
ORDER BY total_potfis DESC"""
fazer_query(query_1)
```

```python
# Descobre a porcentagem de usinas multi-estado que também são multi-município
# Esse número nunca deve ser diferente de 100, pois implicaria em um município pertencer a 2 estados
# Obter 100% indica que o dataset, pelo menos neste aspecto, não é horrível
query_2 = """
    WITH multi_state AS (
        SELECT u.id
        FROM ava1_schema."Usina" u
        JOIN ava1_schema."Localizacoes" l ON u.id_localizacao = l.id_localização
        JOIN ava1_schema."Municipios" m1 ON l.id_municipio1 = m1.id_municipio
        JOIN ava1_schema."Municipios" m2 ON l.id_municipio2 = m2.id_municipio
        WHERE l.id_municipio2 IS NOT NULL
          AND m1.uf <> m2.uf
    )
    
    SELECT
        ROUND(
            100.0 * COUNT(*) FILTER (
                WHERE l.id_municipio2 IS NOT NULL
            ) / NULLIF(COUNT(*), 0),
        2) AS perc_multi_municipio_among_multi_estado
    FROM ava1_schema."Usina" u
    JOIN ava1_schema."Localizacoes" l ON u.id_localizacao = l.id_localização
    WHERE u.id IN (SELECT id FROM multi_state);
"""
fazer_query(query_2)
```

```python
# Encontra as empresas ou proprietários com usinas hidrelétricas em mais de um estado
query_3 = """
    SELECT 
        p.proprietario,
        COUNT(DISTINCT m.uf) AS num_estados
    FROM ava1_schema."Usina" u
    JOIN ava1_schema."Proprietarios" p ON u.id_proprietario = p.id_proprietario
    JOIN ava1_schema."Localizacoes" l ON u.id_localizacao = l.id_localização
    JOIN ava1_schema."Municipios" m ON l.id_municipio1 = m.id_municipio
    GROUP BY p.proprietario
    HAVING COUNT(DISTINCT m.uf) > 1;
"""

fazer_query(query_3);
```

```python
# Retorna os estados com maior e seus números de usinas com mais de um município
query_4 = """
    SELECT m.uf, COUNT(*) AS duplas
    FROM ava1_schema."Localizacoes" l
    JOIN ava1_schema."Municipios" m ON l.id_municipio1 = m.id_municipio
    WHERE l.id_municipio2 IS NOT NULL
    GROUP BY m.uf
    ORDER BY duplas DESC;
"""
fazer_query(query_4)
```

```python
# Retorna a porcentagem de usinas por estado que afetam mais de um município.
query_5 = """
    SELECT 
        est.uf,
        COUNT(*) FILTER (WHERE l.id_municipio2 IS NOT NULL) * 100.0 / COUNT(*) AS perc_duas_municipios
    FROM ava1_schema."Usina" u
    JOIN ava1_schema."Localizacoes" l ON u.id_localizacao = l.id_localização
    JOIN ava1_schema."Municipios" m ON l.id_municipio1 = m.id_municipio
    JOIN ava1_schema."Estados" est ON m.uf = est.uf
    GROUP BY est.uf
    ORDER BY perc_duas_municipios DESC;
"""
fazer_query(query_5)
```
